# Retrieval Augmented Generation
## Step 1: Document Loading
 
- In retrieval augmented generation (RAG), an LLM retrieves contextual documents from an external dataset as part of its execution. 
- This is useful if we want to ask question about specific documents (e.g., our PDFs, words, text, a set of videos, etc).
- This exercise gives us ability to load different document types

### PDFs

- Let's download a sample PDF to data folder. These documents are the result of automated transcription so words and sentences are sometimes split unexpectedly.
- Use **PyPDFLoader**

```python
import os
os.makedirs('data', exist_ok=True)

# Download the PDF document used in this notebook
!wget -O data/Understanding_Climate_Change.pdf https://raw.githubusercontent.com/NirDiamant/RAG_TECHNIQUES/main/data/Understanding_Climate_Change.pdf
path = "data/Understanding_Climate_Change.pdf"
```

```python
path = "data/Understanding_Climate_Change.pdf"
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader(path)
pages = loader.load()
```

Each page is a `Document`.

A `Document` contains text (`page_content`) and `metadata`.

```python
len(pages)
```

```python
page = pages[0]
```

```python
print(page.page_content[:1000])
```

```python
page.metadata
```

### Have a folder full of PDFs?
- In mdocs folder, there are multiple PDFs
- Read all PDFs and merge into 1 large one

```python
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader   
```

```python
loader = DirectoryLoader(
    "mdocs",
    glob="*.pdf",          # match PDFs (use ** for recursion)
    loader_cls=PyPDFLoader,   # use PyPDF for each matched file
    show_progress=True,
    use_multithreading=True,
    max_concurrency=8,
)

docs = loader.load()
print(len(docs), docs[0].metadata)
```
