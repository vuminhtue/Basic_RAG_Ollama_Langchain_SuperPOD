## Step 2: Document Splitting
- There exists limitation of token lenght that LLMs can process, thus, we have to split the long text into smaller chunks
  
<img src="img/RAG.JPG" width="2000"/>

|Text Splitter |	Best  Use Case |	Pros	| Cons
|---|---|---|---|
|**CharacterTextSplitter**	|Simple text, logs	|Fast, simple	|Breaks sentences & meaning|
|**RecursiveCharacterTextSplitter** ⭐	|PDFs, docs, RAG	|Preserves structure, robust	|Slightly slower|
|**TokenTextSplitter**	|Strict token limits	|Exact token control	|Ignores semantic boundaries|
|RecursiveCharacter + tiktoken ⭐	|OpenAI RAG	Token-safe + structured	OpenAI-centric
|MarkdownTextSplitter	|Markdown docs	|Respects headers	|Poor for plain text|
|HTMLHeaderTextSplitter	|Web pages	|Structural splitting	|Needs clean HTML|
|PythonCodeTextSplitter	|Code RAG	|Syntax-aware	|Python only|
|SentenceTransformersTokenTextSplitter	|SBERT embeddings	|Embedding-optimized	|Less flexible|
|SemanticChunker	|High-quality RAG	|Best semantic coherence	|Slow, costly, complex|

## Recursive splitting details

`RecursiveCharacterTextSplitter` is recommended for generic text. 

```python
some_text = """When writing documents, writers will use document structure to group content. \
This can convey to the reader, which idea's are related. For example, closely related ideas \
are in sentances. Similar ideas are in paragraphs. Paragraphs form a document. \n\n  \
Paragraphs are often delimited with a carriage return or two carriage returns. \
Carriage returns are the "backslash n" you see embedded in this string. \
Sentences have a period at the end, but also, have a space.\
and words are separated by space."""
```

```python
len(some_text)
```

```python
c_splitter = CharacterTextSplitter(
    chunk_size=450,
    chunk_overlap=0,
    separator = ' '
)
r_splitter = RecursiveCharacterTextSplitter(
    chunk_size=450,
    chunk_overlap=0, 
    separators=["\n\n", "\n", " ", ""]
)
```

```python
c_splitter.split_text(some_text)
```

```python
r_splitter.split_text(some_text)
```

Let's reduce the chunk size a bit and add a period to our separators:

```python
r_splitter = RecursiveCharacterTextSplitter(
    chunk_size=150,
    chunk_overlap=0,
    separators=["\n\n", "\n", "\. ", " ", ""]
)
r_splitter.split_text(some_text)
```

```python
r_splitter = RecursiveCharacterTextSplitter(
    chunk_size=150,
    chunk_overlap=0,
    separators=["\n\n", "\n", "(?<=\. )", " ", ""]
)
r_splitter.split_text(some_text)
```

## Token splitting

We can also split on token count explicity, if we want.

This can be useful because LLMs often have context windows designated in tokens.

Tokens are often ~4 characters.

```python
from langchain_text_splitters import TokenTextSplitter
```

```python
text_splitter = TokenTextSplitter(chunk_size=1, chunk_overlap=0)
```

```python
text1 = "foo bar bazzyfoo"
```

```python
text_splitter.split_text(text1)
```

```python
text_splitter = TokenTextSplitter(chunk_size=10, chunk_overlap=0)
```

## Recommendation

For most RAG use cases, **RecursiveCharacterTextSplitter** is the best default choice. It produces the most semantically coherent chunks because it always tries to split at the most natural boundary possible.
